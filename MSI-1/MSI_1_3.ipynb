```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('./datasets/StudentsPerformance.csv')
```

```python
df
```

```python
df.describe()
```

```python
plt.plot(df['math score'],'o',markersize=4)
plt.xlabel('Student')
plt.ylabel('Math Score')
plt.title('Math Score of Students')
```

```python
plt.plot(df['math score'],'x', label='math',markersize=3)
plt.plot(df['reading score'],'o', label='reading',markersize=3)
plt.plot(df['writing score'],'v', label='writing',markersize=3)
plt.xlabel('Student')
plt.ylabel('Score')
plt.title('Scores of Students')
plt.legend()
```

```python
plt.subplot(3,1,1)
plt.plot(df['math score'],'bx', label='math',markersize=3)
plt.ylabel('Score')
plt.legend()
plt.subplot(3,1,2)
plt.plot(df['reading score'],'ro', label='reading',markersize=3)
plt.ylabel('Score')
plt.legend()
plt.subplot(3,1,3)
plt.plot(df['writing score'],'gv', label='writing',markersize=3)
plt.xlabel('Student')
plt.ylabel('Score')
plt.tight_layout()
plt.legend()
```

```python
plt.subplot(1,3,1)
plt.plot(df['math score'],'bx', label='math',markersize=3)
plt.ylabel('Score')
plt.legend()
plt.subplot(1,3,2)
plt.plot(df['reading score'],'ro', label='reading',markersize=3)
plt.ylabel('Score')
plt.legend()
plt.subplot(1,3,3)
plt.plot(df['writing score'],'gv', label='writing',markersize=3)
plt.xlabel('Student')
plt.ylabel('Score')
plt.legend()
```
