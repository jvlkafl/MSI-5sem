```python
df_SVC=pd.read_csv('data.csv', sep=';')
```

```python
X = df_SVC.drop(['Target'], axis=1)
y = df_SVC['Target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
import sklearn.metrics as m
from sklearn import preprocessing as p
data_ss = p.StandardScaler().fit_transform(X)
data_ss
```

```python
def classification(X,y, kernel, train_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = train_size, random_state=42)
    model = SVC(kernel = kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = m.accuracy_score(y_test, y_pred)
    prec = m.precision_score(y_test, y_pred, average='weighted')
    rec = m.recall_score(y_test, y_pred, average='weighted')
    f1 = m.f1_score(y_test, y_pred, average='weighted')
    return round(acc, 4)

results = []
kernels = ['linear', 'poly', 'rbf', 'sigmoid']
train_set_sizes = [0.2, 0.4, 0.6, 0.8]
for kernel in kernels:
    small_list = []
    for size in train_set_sizes:
        small_list.append(classification(data_ss,y, kernel, size))
        print(kernel, ',', size, ',', small_list[-1])
    results.append(small_list)
            
```

```python
def classification(X,y, kernel, train_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = train_size, random_state=42)
    model = SVC(kernel = kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = m.accuracy_score(y_test, y_pred)
    prec = m.precision_score(y_test, y_pred, average='weighted')
    rec = m.recall_score(y_test, y_pred, average='weighted')
    f1 = m.f1_score(y_test, y_pred, average='weighted')
    return acc, prec, rec, f1

results = []
kernels = ['linear', 'rbf']
train_set_sizes = [0.2, 0.4, 0.6, 0.8]
for kernel in kernels:
    small_list = []
    for size in train_set_sizes:
        small_list.append(classification(data_ss,y, kernel, size))
        print(kernel, ',', size, ',', small_list[-1])
    results.append(small_list)
            
```

```python

for i in range(len(kernels)):
    plt.plot(train_set_sizes, results[1][i],label=kernels[i])
    plt.xlabel("test set size")
    plt.ylabel("accuracy")
plt.legend()
```

```python
def clas_linear(X,y, c_value):
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=42) #zmienić random_state
    model = SVC(kernel ='rbf', C = c_value)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = m.accuracy_score(y_test, y_pred)
    prec = m.precision_score(y_test, y_pred, average='macro')
    rec = m.recall_score(y_test, y_pred, average='macro')
    f1 = m.f1_score(y_test, y_pred, average='macro')
    return round(acc,3)
```

```python
results = []
c_values = [0.001, 0.01, 0.1, 1.0, 10]#tutaj umieść listę wartości do sprawdzenia
for c in c_values:
        results.append(clas_linear(X, y, c))
        print('dla c=', c,':', results[-1])
```
